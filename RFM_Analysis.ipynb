#**RFM Analysis**

RFM adalah singkatan dari Recency, Frequency, dan Monetary Value, yang masing-masing merupakan indikator penting dalam memahami perilaku pelanggan.
* Recency (Kedekatan Waktu): Mengukur seberapa baru seorang pelanggan melakukan pembelian. Pelanggan yang baru saja berbelanja cenderung lebih responsif terhadap promosi dan komunikasi dari bisnis.
* Frequency (Frekuensi): Mengukur seberapa sering pelanggan melakukan pembelian. Pelanggan yang berbelanja lebih sering menunjukkan tingkat keterlibatan dan kepuasan yang lebih tinggi terhadap merek.
* Monetary Value (Nilai Transaksi): Mengukur jumlah uang yang dibelanjakan oleh pelanggan dalam setiap transaksi. Meskipun setiap pembelian memiliki nilai, menggabungkan faktor ini dengan seberapa baru dan sering pelanggan bertransaksi dapat memberikan gambaran apakah mereka adalah pelanggan yang setia terhadap merek

RFM dapat membantu Anda menjawab pertanyaan-pertanyaan berikut:

1. **Kepada siapa Perusahaan dapat menawarkan produk bernilai lebih tinggi?** Identifikasi pelanggan yang cenderung membeli lebih banyak atau produk dengan harga lebih tinggi berdasarkan pola pembelian mereka.

2. **Kepada siapa Perusahaan dapat mengirim pesan personal untuk mencegah mereka berhenti berbelanja?** Temukan pelanggan yang berisiko tidak kembali (churn) dan kirimkan penawaran atau komunikasi yang relevan untuk mempertahankan mereka.

3. **Siapa yang paling mungkin berinteraksi dengan merek?** Kenali pelanggan yang paling terlibat dan memiliki potensi besar untuk merespons promosi atau kampanye.

4. **Kepada siapa Perusahaan dapat menawarkan promosi dengan waktu terbatas untuk membuat pelanggan berbelanja lagi?** Identifikasi pelanggan yang sudah lama tidak bertransaksi dan kirimkan penawaran eksklusif untuk menarik perhatian mereka kembali.

Analisis RFM mengelompokkan semua pelanggan ke dalam sepuluh grup berdasarkan skor mereka. Perusahaan juga dapat membuat Audience untuk setiap grup guna mengambil tindakan lanjutan atau menggali lebih dalam ke dalam analisis.


Setiap pelanggan akan diberikan segmen RFM berdasarkan skor mereka (dengan nilai bulat 1 hingga 5) pada tiga matrik berikut:
1. Recency (Kedekatan Waktu): Jumlah hari sejak pembelian terakhir dilakukan oleh pelanggan.
2. Frequency (Frekuensi): Jumlah total pesanan yang dilakukan oleh pelanggan.
3. Monetary (Nilai Transaksi): Pendapatan seumur hidup pelanggan (Lifetime Revenue/LTR).

Pendekatan ini membantu bisnis memahami perilaku pelanggan secara lebih mendalam dan memungkinkan tindakan yang lebih strategis berdasarkan masing-masing segmen.


RFM Analysis adalah alat yang sangat berguna dalam berbagai aspek bisnis, terutama dalam memahami dan mengelola hubungan pelanggan. Berikut adalah beberapa penggunaan utama dari RFM Analysis:
1. Segmentasi Pelanggan

Mengelompokkan pelanggan berdasarkan perilaku pembelian mereka (Recency, Frequency, Monetary).
Membuat segmen seperti:
* Champions: Pelanggan terbaik dengan skor tinggi di semua dimensi.
* At Risk: Pelanggan yang sudah lama tidak bertransaksi.
* Hibernating: Pelanggan yang jarang atau tidak aktif.

Membantu fokus pada segmen yang paling berharga atau memerlukan perhatian khusus.

2. Peningkatan Loyalitas Pelanggan
* Identifikasi pelanggan setia: Fokus pada pelanggan dengan skor Frequency dan Monetary tinggi.
* Kampanye personal: Tawarkan diskon, akses eksklusif, atau program loyalitas untuk pelanggan berharga.

3. Retensi Pelanggan
* Mengidentifikasi pelanggan yang berisiko churn (At Risk) berdasarkan skor Recency yang rendah.
* Meluncurkan kampanye "We Miss You" atau memberikan insentif untuk mengaktifkan kembali pelanggan yang sudah lama tidak bertransaksi.

4. Optimasi Pemasaran

Menyesuaikan strategi pemasaran berdasarkan segmen pelanggan:
* Promosi eksklusif untuk Champions.
* Penawaran waktu terbatas untuk Potential Loyalists.
* Reminder atau diskon besar untuk At Risk dan Hibernating.

Meningkatkan ROI dengan fokus pada segmen pelanggan yang memberikan nilai tertinggi.

5. Peningkatan Penjualan
* Upselling: Tawarkan produk premium kepada pelanggan dengan skor Monetary tinggi.
* Cross-selling: Rekomendasikan produk terkait kepada pelanggan dengan skor Frequency tinggi.
* Fokus pada pelanggan yang cenderung membeli lebih banyak atau lebih sering.

6. Perencanaan Produk dan Pengembangan Layanan
* Menganalisis pola pembelian untuk memahami preferensi pelanggan.
* Merancang produk atau layanan baru yang sesuai dengan kebutuhan pelanggan dari segmen tertentu.

7. Evaluasi dan Pelaporan Performa
* Memantau perubahan skor RFM dari waktu ke waktu untuk menilai efektivitas strategi pemasaran.
* Melihat bagaimana pelanggan berpindah antar segmen, misalnya dari Promising ke Champions atau dari Champions ke At Risk.

8. Prediksi Perilaku Pelanggan
* Menggunakan skor RFM untuk memprediksi kemungkinan pelanggan melakukan pembelian lagi.
* Menentukan pelanggan mana yang memiliki potensi untuk menjadi pelanggan setia atau pelanggan berisiko churn.

9. Penyesuaian Anggaran Pemasaran
* Alokasikan anggaran lebih banyak pada pelanggan bernilai tinggi (Frequency dan Monetary tinggi).
* Kurangi investasi pemasaran pada pelanggan yang tidak aktif atau bernilai rendah kecuali untuk kampanye pemulihan khusus.

RFM Analysis adalah alat yang sangat fleksibel dan dapat disesuaikan untuk berbagai tujuan bisnis, membantu perusahaan meningkatkan pendapatan, loyalitas pelanggan, dan efisiensi pemasaran.

**1. Siapa pelanggan bernilai tinggi?**

```python
import pandas as pd
from datetime import datetime
from IPython.display import display

def display_dataframe_to_user(name, dataframe):
    """
    Menampilkan pandas DataFrame kepada pengguna dengan nama yang diberikan.

    Args:
        name (str): Nama yang akan ditampilkan dengan DataFrame.
        dataframe (pd.DataFrame): pandas DataFrame yang akan ditampilkan.
    """
    print(f"{name}:\n")
    display(dataframe)

# Hitung Recency
current_date = datetime.now()
dataset['Last_Purchase_Date'] = pd.to_datetime(dataset['Last_Purchase_Date'])
dataset['Recency'] = (current_date - dataset['Last_Purchase_Date']).dt.days

# Buat RFM Scores
r_bins = pd.qcut(dataset['Recency'], q=5, labels=[5, 4, 3, 2, 1])
f_bins = pd.qcut(dataset['Total_Purchases'], q=5, labels=[1, 2, 3, 4, 5])
m_bins = pd.qcut(dataset['Lifetime_Value'], q=5, labels=[1, 2, 3, 4, 5])

dataset['R_Score'] = r_bins.astype(int)
dataset['F_Score'] = f_bins.astype(int)
dataset['M_Score'] = m_bins.astype(int)

# Gabungkan RFM Scores
dataset['RFM_Score'] = dataset['R_Score'].astype(str) + dataset['F_Score'].astype(str) + dataset['M_Score'].astype(str)

# Identifikasi High-Value Customers
high_value_customers = dataset[dataset['RFM_Score'] == "555"]

# Tampilkan High-Value Customers
display_dataframe_to_user(name="High-Value Customers (RFM Score: 555)", dataframe=high_value_customers)
```

Berikut adalah pelanggan bernilai tinggi (dengan RFM Score: 555), yang menunjukkan bahwa mereka:

1. Baru saja melakukan pembelian.
2. Sering berbelanja.
3. Memiliki total nilai pembelian yang tinggi.

```python
import matplotlib.pyplot as plt

# Ekstraksi pelanggan bernilai tinggi untuk visualisasi
high_value_data = high_value_customers[['Recency', 'Total_Purchases', 'Lifetime_Value']]

# Membuat plot distribusi Recency, Frequency, dan Monetary
plt.figure(figsize=(12, 8))

# Recency
plt.subplot(3, 1, 1)
plt.hist(high_value_data['Recency'], bins=10, alpha=0.7, edgecolor='black', color='#98FB98')
plt.title('Distribusi Recency (Pelanggan Bernilai Tinggi)', fontsize=14)
plt.xlabel('Recency (Hari)', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)

# Frequency
plt.subplot(3, 1, 2)
plt.hist(high_value_data['Total_Purchases'], bins=10, alpha=0.7, edgecolor='black', color='#98FB98')
plt.title('Distribusi Frequency (Total Pembelian)', fontsize=14)
plt.xlabel('Total Pembelian', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)

# Monetary
plt.subplot(3, 1, 3)
plt.hist(high_value_data['Lifetime_Value'], bins=10, alpha=0.7, edgecolor='black', color='#98FB98')
plt.title('Distribusi Monetary (Nilai Seumur Hidup)', fontsize=14)
plt.xlabel('Nilai Seumur Hidup ($)', fontsize=12)
plt.ylabel('Jumlah Pelanggan', fontsize=12)

plt.tight_layout()
plt.show()
```

**2. Siapa pelanggan yang berisiko churn?**

**Kriteria Pelanggan yang Berisiko Churn**

**1. Recency Rendah (R = 1 atau 2)**

Alasan:
* Pelanggan yang sudah lama tidak melakukan pembelian menunjukkan penurunan keterlibatan dengan merek.
* Dalam bisnis, semakin lama pelanggan tidak berbelanja, semakin besar kemungkinan mereka telah beralih ke pesaing atau kehilangan minat.

Mengapa Penting?
* Dimensi Recency adalah indikator paling langsung untuk mengukur risiko churn. Pelanggan dengan Recency rendah membutuhkan upaya untuk diaktifkan kembali sebelum churn menjadi permanen.

**2. Frequency Rendah (F = 1 atau 2)**

Alasan:
* Pelanggan dengan frekuensi pembelian rendah menunjukkan hubungan yang lemah dengan bisnis.
* Mereka cenderung hanya melakukan pembelian sesekali atau sporadis, sehingga lebih rentan untuk churn dibandingkan pelanggan dengan frekuensi tinggi.

Mengapa Penting?
* Frequency rendah menunjukkan kurangnya pola keterlibatan yang konsisten. Jika pelanggan seperti ini tidak diberi perhatian, mereka cenderung berhenti sepenuhnya.

**3. Monetary (M Tinggi atau Rendah)**

Alasan:
* Pelanggan dengan Monetary rendah tidak memberikan kontribusi finansial yang signifikan, sehingga prioritas untuk mengaktifkan mereka kembali lebih rendah.
* Pelanggan dengan Monetary tinggi adalah pelanggan bernilai tinggi yang telah memberikan kontribusi besar, tetapi jika Recency dan Frequency mereka rendah, mereka berisiko hilang sebagai aset finansial penting.

Mengapa Penting?
* Fokus pada Monetary tinggi memungkinkan perusahaan memprioritaskan sumber daya untuk pelanggan yang dapat memberikan dampak terbesar pada pendapatan.

**Kombinasi untuk Identifikasi Pelanggan Berisiko Churn**
* R Rendah (R = 1 atau 2): Pelanggan sudah lama tidak aktif.
* F Rendah (F = 1 atau 2): Pelanggan jarang melakukan pembelian.
* Prioritaskan M Tinggi: Untuk pelanggan bernilai besar yang perlu diaktifkan kembali dengan cepat.

**Kesimpulan**

Kriteria ini digunakan karena:

1. Mereka memberikan indikator langsung tentang ketidakaktifan pelanggan.
2. Dimensi Recency adalah sinyal awal potensi churn, sementara Frequency dan Monetary membantu memprioritaskan tindakan.
3. Analisis ini memungkinkan perusahaan mengidentifikasi pelanggan yang berisiko sebelum mereka benar-benar hilang, sehingga langkah intervensi (seperti kampanye retensi) dapat diambil secara proaktif.

```python
def display_dataframe_to_user(name, dataframe):
    """
    Menampilkan pandas DataFrame kepada pengguna dengan nama yang diberikan.

    Args:
        name (str): Nama yang akan ditampilkan dengan DataFrame.
        dataframe (pd.DataFrame): pandas DataFrame yang akan ditampilkan.
    """
    print(f"{name}:\n")
    display(dataframe)

# Define criteria for customers at risk of churn
# R = 1 or 2 (low Recency), F = 1 or 2 (low Frequency)
churn_risk_customers = dataset[(dataset['R_Score'] <= 2) & (dataset['F_Score'] <= 2)]

# Sort the result by Monetary Score to prioritize high-value customers at risk
churn_risk_customers = churn_risk_customers.sort_values(by='M_Score', ascending=False)

# Display the results to the user
display_dataframe_to_user(name="Customers at Risk of Churn", dataframe=churn_risk_customers)
```

**3. Segmen mana yang paling responsif terhadap promosi?**

Kriteria untuk Segmen Responsif terhadap Promosi

**1. Recency Tinggi (R = 4 atau 5)**

Pelanggan yang baru-baru ini melakukan pembelian lebih cenderung responsif terhadap promosi karena mereka masih memiliki hubungan aktif dengan bisnis.

**2. Frequency Tinggi (F = 4 atau 5)**

Pelanggan yang sering melakukan pembelian menunjukkan keterlibatan yang kuat, sehingga mereka lebih mungkin merespons promosi.

**3. Monetary Bervariasi**

Pelanggan dengan nilai Monetary tinggi memberikan potensi dampak finansial lebih besar ketika mereka merespons promosi.

**Alasan Kriteria Tersebut**
1. Recency Tinggi: Pelanggan yang baru saja berbelanja lebih mudah terhubung dengan merek karena mereka masih memiliki pengalaman belanja yang segar.
2. Frequency Tinggi: Pelanggan yang sering berbelanja cenderung memiliki pola keterlibatan yang stabil dan lebih mungkin terpengaruh oleh penawaran.
3. Monetary Bervariasi: Pelanggan dengan Monetary tinggi sering kali dianggap sebagai prioritas, tetapi Monetary yang rendah atau sedang tetap penting untuk diaktifkan kembali atau diubah menjadi pelanggan bernilai tinggi.

```python
# Menentukan kriteria pelanggan yang paling responsif terhadap promosi
# R = 4 atau 5 (Recency tinggi), F = 4 atau 5 (Frequency tinggi)
pelanggan_responsif = dataset[(dataset['R_Score'] >= 4) & (dataset['F_Score'] >= 4)]

# Mengurutkan berdasarkan Monetary Score untuk memprioritaskan pelanggan bernilai tinggi
pelanggan_responsif = pelanggan_responsif.sort_values(by='M_Score', ascending=False)

# Menampilkan hasil kepada pengguna
display_dataframe_to_user(name="Pelanggan Paling Responsif terhadap Promosi", dataframe=pelanggan_responsif)
```

```python
# Mendefinisikan segmen pelanggan berdasarkan skor RFM
def segmen_pelanggan(baris):
    if baris['R_Score'] == 5 and baris['F_Score'] == 5 and baris['M_Score'] == 5:
        return 'Juara (Champions)'
    elif baris['R_Score'] >= 4 and baris['F_Score'] >= 4:
        return 'Pelanggan Setia (Loyal Customers)'
    elif baris['R_Score'] <= 2 and baris['F_Score'] >= 4:
        return 'Berisiko (At Risk)'
    elif baris['R_Score'] <= 2 and baris['F_Score'] <= 2:
        return 'Hibernasi (Hibernating)'
    elif baris['R_Score'] >= 3 and baris['F_Score'] <= 2:
        return 'Menjanjikan (Promising)'
    else:
        return 'Lainnya (Others)'

# Menerapkan fungsi untuk mengelompokkan pelanggan
dataset['Segmen'] = dataset.apply(segmen_pelanggan, axis=1)

# Menghitung jumlah pelanggan di setiap segmen
jumlah_segmen = dataset['Segmen'].value_counts().reset_index()
jumlah_segmen.columns = ['Segmen', 'Jumlah Pelanggan']

# Menampilkan hasil segmentasi kepada pengguna
display_dataframe_to_user(name="Hasil Segmentasi Pelanggan", dataframe=jumlah_segmen)
```

Segmentasi pelanggan berdasarkan analisis RFM telah selesai. Berikut adalah hasilnya:

1. Champions: Pelanggan terbaik dengan skor tertinggi di semua dimensi.
2. Loyal Customers: Pelanggan setia yang sering dan baru saja berbelanja.
3. At Risk: Pelanggan bernilai tinggi tetapi sudah lama tidak aktif.
4. Hibernating: Pelanggan jarang aktif dan berisiko tinggi churn.
5. Promising: Pelanggan dengan potensi untuk menjadi pelanggan bernilai tinggi.
